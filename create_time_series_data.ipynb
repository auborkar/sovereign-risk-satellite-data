```python
import matplotlib.pylab as plt
import geopandas as gp
import rioxarray as rxr
import pandas as pd
import numpy as np
import datetime

from datetime import date


import requests # request img from web
import shutil 

import os 
import urllib.request

from os.path import expanduser as base_dir


from os import listdir
from os.path import isfile, join
```

```python
def open_tif(tif_path):
    return rxr.open_rasterio(tif_path)


def calc_mean_esi_region(tif_name, tif_path, argn_sf):
    tif = open_tif(tif_path)
    date = tif_name.split('_')[-1].split('.')[0]
    tif_date = datetime.datetime.strptime(date, '%Y%j').strftime('%Y%m%d')

    
    ds_masked = tif.where(tif!=-9999.)
    results = pd.DataFrame(columns = ['date', 'ADM1_PCODE', 'mean_esi'])
    
    for ADM1_PCODE, country_geom in zip(argn_sf.ADM1_PCODE, argn_sf.geometry): 
        clipped = ds_masked.rio.clip([country_geom])
        mean_esi = clipped.mean().values.tolist()
        results = results.append({'ADM1_PCODE': ADM1_PCODE, 'mean_esi': mean_esi}, ignore_index=True)

    results['date'] = [tif_date for i in range(len(results)) ]

    return results
```

```python
argn_sf = gp.read_file(base_dir('~/sovereign-risk-satellite-data/Data/arg_adm_unhcr2017_shp'))
argn_sf = argn_sf.to_crs('WGS 1984')
```

```python
tracker_df = pd.DataFrame(columns = ['tif_name', 'processed?', 'date' ])

time_series = pd.DataFrame(columns = ['date', 'ADM1_PCODE', 'mean_esi'])

error_df = pd.DataFrame(columns = ['tif_name', 'error?', 'date' ])
```

```python
# process the tifs
tif_folder = '/Users/gracecolverd/sovereign-risk-satellite-data/Data/tifs'
avail_tif_names  = [ f for f in listdir(tif_folder) if ((isfile(join(tif_folder, f))) &  (f[-4:] == '.tif') ) ]
avail_tif_paths = [os.path.join( tif_folder, f ) for f in avail_tif_names]

for tif_nam, tif_path in zip(avail_tif_names, avail_tif_paths):
    print(tif_nam, tif_path)
    
    if len(tracker_df[tracker_df['tif_name'] == tif_nam]['processed?'].values ) ==0 :
        try:
            results = calc_mean_esi_region(tif_nam, tif_path, argn_sf)
            if len(results) == 24:
                time_series= time_series.append(results)
                tracker_df = tracker_df.append({'tif_name': tif_nam, 'processed?' :'Y', 'date' : date.today().strftime('%Y%m%d')}, ignore_index=True)
                print('finished {}'.format(tif_nam))  
            else:
                print('results length incorrect - check {}'.format(tif_nam))
        except:
           
            if len(error_df[error_df['tif_name'] == tif_nam]['error?'].values ) ==0 : 
                error_df = error_df.append({'tif_name': tif_nam, 'error?' :'Y', 'date' : date.today().strftime('%Y%m%d')}, ignore_index=True)
            
    else:
        print('already processed')
```

```python
final = time_series.set_index('ADM1_PCODE').join(argn_sf.set_index('ADM1_PCODE'))
```
